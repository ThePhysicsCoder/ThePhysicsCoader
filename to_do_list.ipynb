```python
# to do list code

tasks=[]


def show_menu():
  print("\n_______to do list menu_______\n")
  print("select your options:\n")
  print("1.add tasks")
  print("2.view tasks")
  print("3.mark the task as done")
  print("4.save tasks to file")
  print("5.delete a task")
  print("6.load tasks from file")
  print("7.exit\n")



def add_tasks():
  try:
    task=input("enter the task:")
    tasks.append(task)
    print("task added successfully")
    user=input("want to add more? (yes/no): ")
    if user.lower()=="yes":
      while True:
        task=input("enter the task (type no to stop): ")
        if task.lower()=="no":
          print("thank you!")
          break
        tasks.append(task)
        print("task added successfully\n")
    elif user.lower()=="no":
      pass # No need to break here, as the loop condition is based on user input inside the loop.
  except ValueError:
    print("invalid input")

def view_tasks():
  print('tasks:')
  for task in range (len(tasks)):
    print(f"{task+1}.{tasks[task]}")
  print("\n")


y=": done"

def mark_done():
  try:
    number=int(input("enter the number of the task you want to mark done: "))
    tasks[number-1]=tasks[number-1]+ y
    print("task marked done successfully\n")
  except ValueError:
    print("invalid input")


def save_tasks():
  with open('tasks.txt', 'w') as file:
    for task in tasks:
      file.write(task+ "\n")
  print("tasks saved successfully\n")


def load_task():
  with open('tasks.txt', 'r') as file:
    lines=file.readlines()
    for line in lines:
      tasks.append(line.strip())
  print("tasks loaded successfully\n")


def delete_task():
  number=int(input("enter the number of the task you want to delete: "))
  delete=tasks.pop(number-1)
  print(f"task '{delete}' deleted successfully\n")


#main art of coding

def showing():
  while True:
    show_menu()
    choice=int(input("enter your choice: "))
    if choice==1:
      add_tasks()
    elif choice==2:
      view_tasks()
    elif choice==3:
      mark_done()
    elif choice==4:
      save_tasks()
    elif choice==5:
      delete_task()
    elif choice==6:
      load_task()
    elif choice==7:
      print("thank you for using our program")
      break
    else:
      print("invalid choice")
showing()
```
